```python
def split_name(df,col_name,new_name):
    
    split_arr = df[col_name].str.split(',')
    result = []
    for i in range(len(split_arr)):
         result.append(split_arr[i][1].split('.')[0].replace(' ',''))
    df[new_name] = result
    return df
```
